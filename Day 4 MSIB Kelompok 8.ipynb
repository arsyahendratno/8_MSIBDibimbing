```python
class BangunDatar():
  Semua_Luas = []

  def lpersegi(self):
    print('Mari Hitung Luas Persegi Kamu Yuk!')
    self.sisi= int(input('Sisi = '))
    self.luas = (self.sisi * self.sisi)
    print(f"{self.sisi} * {self.sisi} = {self.luas}\n")
    BangunDatar.Semua_Luas.append(self.luas)

  def lpersegi_panjang(self):
    print('Mari Hitung Luas Persegi Panjang Kamu Yuk!')
    self.panjang= int(input('Panjang = '))
    self.lebar= int(input('Lebar = '))
    self.luas = (self.panjang * self.lebar)
    print(f"{self.panjang} * {self.lebar} = {self.luas}\n")
    BangunDatar.Semua_Luas.append(self.luas)

  def lsegitiga(self):
    print('Mari Hitung Luas Segitiga Kamu Yuk!')
    self.alas= int(input('Alas = '))
    self.tinggi= int(input('Tinggi = '))
    self.luas = ((self.alas * self.tinggi)/2)
    print(f"({self.alas} * {self.tinggi})/2 = {self.luas}\n")
    BangunDatar.Semua_Luas.append(self.luas)

  def llingkaran(self):
    print('Mari Hitung Luas Lingkaran Kamu Yuk!')
    self.jari= int(input('Jari-jari = '))
    self.luas = (3.14 * self.jari * self.jari)
    print(f"3.14 * {self.jari} * {self.jari} = {self.luas}\n")
    BangunDatar.Semua_Luas.append(self.luas)

  def ltrapesium(self):
    print('Mari Hitung Luas Trapesium Kamu Yuk!')
    self.sjj1= int(input('Sisi Sejajar 1 = '))
    self.sjj2= int(input('Sisi Sejajar 2 = '))
    self.tinggi= int(input('Tinggi = '))
    self.luas = ((self.tinggi*(self.sjj1 + self.sjj2))/2)
    print(f"({self.tinggi}*({self.sjj1} + {self.sjj2}))/2 = {self.luas}\n")
    BangunDatar.Semua_Luas.append(self.luas)

class persegi(BangunDatar):
  pass
class persegi_panjang(BangunDatar):
  pass
class segitiga(BangunDatar):
  pass
class lingkaran(BangunDatar):
  pass
class trapesium(BangunDatar):
  pass
```
